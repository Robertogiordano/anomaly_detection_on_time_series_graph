```python
ACCEPTED_MODELS_PATH = 'new_models'
models+=os.listdir(ACCEPTED_MODELS_PATH)

DISCARDED_MODELS_PATH = 'discarded models'
models += os.listdir(DISCARDED_MODELS_PATH)
```

```python
# Create a DataFrame to mark features
data = {feature: [("X" if feature in model else "") for model in models] for feature in features}
data["Model"] = models
df = pd.DataFrame(data)

# Reorder columns to have "Model" as the first column
df = df[["Model"] + features]
```
