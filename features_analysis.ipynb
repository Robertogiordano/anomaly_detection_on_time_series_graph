```python
import pandas as pd
import os

# Define the list of features based on the model names
features = [
    "data_shuffled",
    "subdata_extended",
    "augmented_dataset",

    "temporal",
    "spatial",
    
    "with_gate",
    
    "batch_16",

    "edge_weight",
    
    "batch_norm",
    
    "feature_loss",
    "combined_loss",
    "lambdaEdgeLoss0.5",
    "edge_weight_preservation_loss",
    "pairwise_similarity_loss",
    "L1Loss",
    "MSE_Latent_Loss",
    "MSE_Latent_Loss_alpha_0.3",
    "MSE_Latent_Loss_alpha_0.7",

    "20days",
    "24days",

    "150epochs",

    "deeper_channels",
    "multiScale",
    "single_conv",
    "double_conv",
    "attentionSpatial",
    
    "kernel_3_stride_1_padding_1",
    "cheb_pol_5",
    
    "timewindow_8",
    "timewindow_16",
    "timewindow_24",
    
]

# List of model names
models = []
```

```python
ACCEPTED_MODELS_PATH = 'new_models'
models += sorted(os.listdir(ACCEPTED_MODELS_PATH), key=lambda x: os.path.getctime(os.path.join(ACCEPTED_MODELS_PATH, x)))


DISCARDED_MODELS_PATH = 'discarded models'
models += sorted(os.listdir(DISCARDED_MODELS_PATH), key=lambda x: os.path.getctime(os.path.join(DISCARDED_MODELS_PATH, x)))
```

```python
# Create a DataFrame to mark features
data = {feature: [("X" if feature in model else "") for model in models] for feature in features}
data["Model"] = models
df = pd.DataFrame(data)

# Reorder columns to have "Model" as the first column
df = df[["Model"] + features]
```
